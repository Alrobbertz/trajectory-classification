```python
from data import *
from evaluation import process_data

import pickle
import time

import pandas as pd
import numpy as np
```

```python
# Load the Raw Data from CSV
x_raw, y_raw = load_raw_data(5)
pickle.dump((x_raw, y_raw), open('raw_data.pkl','wb'))
```

```python
#Load Raw Data
x_raw, y_raw = pickle.load(open('raw_data.pkl','rb'))

x_train_p = []
x_train_np = []
x_train_gen = []

y_train = []

# Extract Features
# For each Trajectory
for i in range(len(x_raw)):
    print(f'Processing Trajectory: {i}')
    
    if len(x_raw[i]) > 1:
        # Process Trajectory
        [p, nop, gen] = process_data(x_raw[i])
        
        if p.shape == (2000, 3):
            # Append to training
            x_train_p.append(p)
            x_train_np.append(nop)
            x_train_gen.append(gen)
            y_train.append(y_raw[i])
        else:
            print('Not enough Values to unpack')
    else:
        print('Too Short to Process')


print('Done Processing')
# Print the Shapes
x_train_p = np.array(x_train_p, ndmin=3)
x_train_np = np.array(x_train_np)
x_train_gen = np.array(x_train_gen)
y_train = np.array(y_train)
print(f'x_train_p: {x_train_p.shape}')
print(f'x_train_np: {x_train_np.shape}')
print(f'x_train_gen: {x_train_gen.shape}')
print(f'Label Shape: {y_train.shape}')  

# Save the Data to an object file
pickle.dump((x_train_p, x_train_np, x_train_gen, y_train), open('train_data.pkl','wb'))
```
