```python
min_lat = df.latitude.min()
max_lat = df.latitude.max()
delta_lat = df.latitude.max() - df.latitude.min()
min_long = df.longitude.min()
max_long = df.longitude.max()
delta_long = df.longitude.max() - df.longitude.min()

scale = 0.5

frame_bot = min_lat - scale * delta_lat
frame_top = max_lat + scale * delta_lat
frame_left = min_long - scale * delta_long
frame_right = max_long + scale * delta_long
frame_height = frame_top - frame_bot
frame_width = frame_right - frame_left

print(f'Min Lat: {min_lat}')
print(f'Max Lat: {max_lat}')
print(f'Delta Lat: {delta_lat}')
print(f'Frame Height: {frame_height}')
print(f'Min Long: {min_long}')
print(f'Min Long: {max_long}')
print(f'Delta Long: {delta_long}')
print(f'Frame Width: {frame_width}')

print(f'\nFrame Bot: {frame_bot}')
print(f'Frame Top: {frame_top}')
print(f'Frame Left: {frame_left}')
print(f'Frame Right: {frame_right}')
```
