```python
import numpy as np
import pickle

from tensorflow.keras.models import load_model

from evaluation import process_data, run

```

```python
# Load Model and Training Data
model = load_model('best_dropout.h5')
x_raw, y_raw = pickle.load(open('raw_data.pkl', 'rb'))

# Save Predictions
predictions = []

i = 0
for test in x_raw:
    if len(test) > 1:
        print(f'Running Test {i}')
        i +=1
        data = process_data(test)
        [p, nop, gen] = data
        if p.shape == (2000, 3):
            result = run(data, model)
            predictions.append(result)
        else:
            print('Not enough Values to unpack')
            print(p.shape)
            predictions.append(-1)
    else:
        print('Too Short')
    
correct = [1 if predictions[i] == y_raw[i] else 0 for i in range(len(predictions))]
print(f'Test Accuracy: {sum(correct)/len(correct)}')
    
```
