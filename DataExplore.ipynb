```python
from data import load_raw_data
from evaluation import process_data_ff
import pickle
import time
import numpy as np


```

```python
#Load Raw Data
x_raw, y_raw = pickle.load(open('raw_data.pkl','rb'))
x_train = []
y_train = []

# Time the Preprocessing 
start_time = time.time()

# Extract Features
# For each Trajectory
for i in range(len(x_raw)):
    print(f'Processing Trajectory: {i}')
    traj_feat = []
    
    # Process Data
    if len(x_raw[i]) > 1:
        # Process Trajectory
        feat = process_data_ff(x_raw[i])
        # Append to training
        x_train.append(feat)
        y_train.append(y_raw[i])  
    else:
        print('Too Short to Process')
    
# Stop the Preprocessing Timer
print('PreProcessing took {} seconds'.format(time.time() - start_time))

x_train = np.array(x_train)
print(x_train.shape)
print(y_raw.shape)

# Save the Data to an object file
pickle.dump((x_train, y_train), open('train_data_ff.pkl','wb'))
```
